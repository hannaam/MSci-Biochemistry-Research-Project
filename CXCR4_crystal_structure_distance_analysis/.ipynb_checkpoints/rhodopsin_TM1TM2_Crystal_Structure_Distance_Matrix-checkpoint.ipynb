```python
file_variable = open('rhodopsin_1N3M_TM1.pdb')
all_lines_variable = file_variable.readlines()

chain=[]
for line in all_lines_variable[100:]:
    if 'CA' in line:
        chain.append(line[17:20])

print(chain)


chain_TM1=['PHE', 'SER', 'MET', 'LEU', 'ALA', 'ALA', 'TYR', 'MET', 'PHE', 'LEU', 'LEU', 'ILE', 'MET', 'LEU', 'GLY', 'PHE', 'PRO', 'ILE', 'ASN', 'PHE', 'LEU', 'THR', 'LEU', 'TYR', 'VAL']
print(len(chain_TM1))
#print(chain)
#print(len(chain))  

file_variable = open('rhodopsin_1N3M_TM2.pdb')
all_lines_variable = file_variable.readlines()

chain=[]
for line in all_lines_variable[100:]:
    if 'CA' in line:
        chain.append(line[17:20])

print(chain)
chain_TM2=['TYR', 'ILE', 'LEU', 'LEU', 'ASN', 'LEU', 'ALA', 'VAL', 'ALA', 'ASP', 'LEU', 'PHE', 'MET', 'VAL', 'PHE', 'GLY', 'GLY', 'PHE', 'THR', 'THR', 'THR', 'LEU', 'TYR']
print(len(chain_TM2))
```

```python
def calc_residue_dist(residueA, residueB) :
    """Returns the C-alpha distance between two residues"""
    diff_vector  = residueA["CA"].coord - residueB["CA"].coord
    return numpy.sqrt(numpy.sum(diff_vector * diff_vector))

calc_residue_dist(residueA,residueB)
print(calc_residue_dist(residueA,residueB))

def calc_dist_matrix(chainA, chainB) :
    """Returns a matrix of C-alpha distances between two chains"""
    answer = numpy.zeros((len(chainA), len(chainB)), numpy.float)
    #print(answer)
    for row, residueA in enumerate(chainA) :
        for col, residueB in enumerate(chainB) :
            answer[row, col] = calc_residue_dist(residueA, residueB)
    return answer
         
calc_dist_matrix(chainA,chainB)
print(calc_dist_matrix(chainA,chainB))
```

```python
dist_matrix = calc_dist_matrix(chainA,chainB)

contact_map = dist_matrix < 10.0

print("Minimum distance", numpy.min(dist_matrix))
print("Maximum distance", numpy.max(dist_matrix))
```

```python
import pandas as pd
df = pd.DataFrame(dist_matrix, index=chain_TM1, columns=chain_TM2)
display(df)
```

```python
df[df > 10] = 0

import matplotlib.pyplot as plt
import seaborn as sns, numpy as np

sns.set(style="white")

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(60, 30))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(df, cmap=cmap,  center=0, vmax=10.0, vmin=0.0, square=True, linewidths=0.5)

# use matplotlib.colorbar.Colorbar object
cbar = ax.collections[0].colorbar
cbar.ax.set_yticklabels(['> 10','2','4','6', '8', '10'],fontsize=30, fontweight='bold')
cbar.set_label('Distance ()', fontsize=30, fontweight='bold')

# here set the labelsize by 20
cbar.ax.tick_params(labelsize=25)
plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')

ax.set_xticklabels(
    ax.get_xticklabels(), 
    rotation=90, 
    horizontalalignment='right',
    fontweight='bold',
    fontsize=20)

ax.set_yticklabels(
    ax.get_yticklabels(),  
    horizontalalignment='right',
    rotation=0,
    fontweight='bold',
    fontsize=20)

cbar.ax.tick_params(labelsize=30)

from matplotlib.patches import Rectangle

ax.set_title('CXCR4 Crystal Structure', fontweight='bold', fontsize=30)
ax.set_xlabel('TM5', fontweight='bold', fontsize=30)
ax.set_ylabel('TM5', fontweight='bold', fontsize=30)

#ax.add_patch(Rectangle((5, 1), 1, 1, fill=False, edgecolor='black', lw=3))

plt.show()
```
