```python
import Bio.PDB
import numpy

p = Bio.PDB.PDBParser()
structureCXCR4_1 = p.get_structure('CXCR4', 'CXCR4_3ODU_TM5_1.pdb')
structureCXCR4_2 = p.get_structure('CXCR4', 'CXCR4_3ODU_TM5_2.pdb')
```

```python
for modelA in structureCXCR4_1:
    print(modelA)
    for chainA in modelA:
        print(chainA)
        for residueA in chainA:
            print(residueA)
            for atomA in residueA:
                print(atomA)
                
for modelB in structureCXCR4_2:
    print(modelB)
    for chainB in modelB:
        print(chainB)
        for residueB in chainB:
            print(residueB)
            for atomB in residueB:
                print(atomB)
```

```python
#Create a list of all the residue 3 letter codes

chain=['VAL', 'VAL', 'VAL', 'PHE', 'GLN', 'PHE', 'GLN', 'HIS', 'ILE', 'MET', 'VAL', 'GLY', 'LEU', 'ILE', 'LEU', 'PRO', 'GLY', 'ILE', 'VAL', 'ILE', 'LEU']
print(chain)
print(len(chain))
```

```python
#THIS CODE WAS TAKEN FROM: https://warwick.ac.uk/fac/sci/moac/people/students/peter_cock/python/protein_contact_map/

def calc_residue_dist(residueA, residueB) :
    """Returns the C-alpha distance between two residues"""
    diff_vector  = residueA["CA"].coord - residueB["CA"].coord
    return numpy.sqrt(numpy.sum(diff_vector * diff_vector))

calc_residue_dist(residueA,residueB)
print(calc_residue_dist(residueA,residueB))

def calc_dist_matrix(chainA, chainB) :
    """Returns a matrix of C-alpha distances between two chains"""
    answer = numpy.zeros((len(chainA), len(chainB)), numpy.float)
    #print(answer)
    for row, residueA in enumerate(chainA) :
        for col, residueB in enumerate(chainB) :
            answer[row, col] = calc_residue_dist(residueA, residueB)
    return answer
         
calc_dist_matrix(chainA,chainB)
print(calc_dist_matrix(chainA,chainB))
```

```python
dist_matrix = calc_dist_matrix(chainA,chainB)

contact_map = dist_matrix < 10.0

print("Minimum distance", numpy.min(dist_matrix))
print("Maximum distance", numpy.max(dist_matrix))
```

```python
import pandas as pd
df = pd.DataFrame(dist_matrix, index=chain, columns=chain)
display(df)
```

```python
df[df > 15] = 0

import matplotlib.pyplot as plt
import seaborn as sns, numpy as np

sns.set(style="white")

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(25, 25))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(df, cmap=cmap,  center=0, vmax=15.0, vmin=0.0, square=True, linewidths=0.5, cbar_kws={"shrink": 0.6})

cbar = ax.collections[0].colorbar
cbar.ax.set_yticklabels(['> 15','2.5','5','7.5', '9', '12.5', '15'])
cbar.set_label('Distance (Å)', fontsize=40, fontweight='bold')
cbar.ax.tick_params(labelsize=30)

ax.set_xticklabels(
    ax.get_xticklabels(), 
    rotation=90, 
    fontsize=40)

ax.set_yticklabels(
    ax.get_yticklabels(),  
    rotation=0,
    fontsize=40)

ax.axhline(y=0, color='k',linewidth=3)
ax.axhline(y=df.shape[0], color='k',linewidth=3)
ax.axvline(x=0, color='k',linewidth=3)
ax.axvline(x=df.shape[0], color='k',linewidth=3)

ax.set_facecolor("whitesmoke")
f.patch.set_alpha(0)

#ax.set_title('Number of Times Residue Pair Interaction Observed (Total=7)', fontweight='bold', fontsize=70,pad=40)
ax.set_xlabel('TM5', fontweight='bold', fontsize=50,labelpad=20)
ax.set_ylabel('TM5', fontweight='bold', fontsize=50,labelpad=20)
ax.invert_yaxis()
#ax.add_patch(Rectangle((5, 1), 1, 1, fill=False, edgecolor='black', lw=3))
plt.savefig('crystal structure CXCR4 TM5TM5',facecolor=f.get_alpha())
plt.show()
```
