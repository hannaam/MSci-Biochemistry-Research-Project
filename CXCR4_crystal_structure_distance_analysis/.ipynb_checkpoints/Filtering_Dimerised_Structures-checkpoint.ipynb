```python
import os
#directory="A2A_TM5TM5/martini_EN/atomistic_files"

chainA=[]
chainA1=[]

chainB=[]
chainB1=[]

distances={} #Distances on their own
distances_residues=[] #Distances with residues names
distance_cutoff={} #Distances with residue names <10 A 
#distance_cutoff=[]
distance_cutoff_all=[]

directory = os.fsencode("A2A_TM5TM5/martini_EN/atomistic_files")

from pathlib import Path
import math

data_folder = Path("A2A_TM5TM5/martini_EN/atomistic_files")

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    file_to_open = data_folder / filename
    print(filename)
    if filename.startswith("TM5TM5"):
        file_to_open = data_folder / filename
        f = open(file_to_open)
        all_lines_variable=f.readlines()
        f.close()
        chainA=[]
        chainA1=[]
        for line in all_lines_variable[3:428]:
            if 'CA' in line:
                chainA.append(line[2:8])
                chainA1.append(line)
            chainB=[]
            chainB1=[]
            for line in all_lines_variable[428:854]:
                if 'CA' in line:
                    chainB.append(line[2:8])
                    chainB1.append(line)
        for chainAline in chainA1:
            x1=float(chainAline[23:28])
            y1=float(chainAline[31:36])
            z1=float(chainAline[39:44])
            for chainBline in chainB1:
                x2=float(chainBline[23:28])
                y2=float(chainBline[31:36])
                z2=float(chainBline[39:44])
                if chainAline != chainBline:
                    d = math.sqrt(math.pow(x2 - x1, 2) +
                    math.pow(y2 - y1, 2) +
                    math.pow(z2 - z1, 2)* 1.0)
                    dist_angstroms=d*10
                    #distances.append(dist_angstroms)
                    #res_dist=[chainAline[5:8],chainBline[5:8],dist_angstroms]
                    res_dist=[chainAline[2:8],chainBline[2:8],dist_angstroms]
                    #distances_residues.append(res_dist)
                    if filename in distances.keys():
                        distances[filename].append(res_dist)
                    else:
                        distances[filename] = [res_dist]
                    if dist_angstroms < 10:
                        #distance_cutoff.append(res_dist) 
                        if filename in distance_cutoff.keys():
                            distance_cutoff[filename].append(res_dist)
                        else:
                            distance_cutoff[filename] = [res_dist]
                        #print(filename, res_dist)
                        #distance_cutoff_all.append{filename, res_dist}
                        #distance_cutoff_all.append(distance_cutoff)

#MAKE DICTIONARY FOR ALL LENGTHS
#MAKE IT VISUALLY APPEALING
        
#print(chainA)
#print(chainA1)
#print(chainB)
#print(chainB1)

#print(x1)
#print(y1)
#print(z1)
#print(x2)
#print(y2)
#print(z2)

#print(distances)
#print(distances_residues)
#print(distance_cutoff)
print(distances)
print(len(distances))
print(distance_cutoff)
print(len(distance_cutoff))
```
